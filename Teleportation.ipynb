# Quantum teleportation

```python
#Transfer quantum information from one to another -> through entanglement resource
from qiskit import *
circuit = QuantumCircuit(3,3)
```

```python
%matplotlib inline
circuit.draw(output='mpl')
```

```python
#begin teleportation preceduce. Transfer qubits from q0 to q2.
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
#First create entanglement between q1 and q2
circuit.h(1)
circuit.cx(1,2)
circuit.draw(output='mpl')
```

```python
# The rest part of the teleportation protocol is to  cx(0,1) and h(0)
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0,1],[0,1])
circuit.draw(output='mpl')
```

```python
#With teleportation protocol
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output='mpl')
```

```python
circuit.measure(2,2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit,backend= simulator,  shots= 1024).result()
counts= result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
